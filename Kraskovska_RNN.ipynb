ùëêùëúùë†(3ùë•) , ùë†ùëñùëõ(ùë• ‚àï 5) , ùë†ùëñùëõ(ùë• + 3) ‚àï 2 —ñ —Ç–¥.

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# –ì–µ–Ω–µ—Ä—É—î–º–æ –¥–∞–Ω—ñ
N = 10000
X = np.random.rand(N) * 10  # –≥–µ–Ω–µ—Ä—É—î–º–æ N —Ç–æ—á–æ–∫ –≤—ñ–¥ 0 –¥–æ 10
Y = np.cos(3*X) + 0.1 * np.random.randn(N)  # –≤—Ö—ñ–¥–Ω—ñ –¥–∞–Ω—ñ - –∫–æ—Å–∏–Ω—É—Å –∑ —à—É–º–æ–º

# –ü–æ–±—É–¥–æ–≤–∞ –º–æ–¥–µ–ª—ñ –Ω–µ–π—Ä–æ–Ω–Ω–æ—ó –º–µ—Ä–µ–∂—ñ
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, input_shape=(1,), activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

# –ö–æ–º–ø—ñ–ª—è—Ü—ñ—è –º–æ–¥–µ–ª—ñ
model.compile(optimizer='adam', loss='mse')

# –ù–∞–≤—á–∞–Ω–Ω—è –º–æ–¥–µ–ª—ñ —Ç–∞ –∑–±–µ—Ä–µ–∂–µ–Ω–Ω—è —ñ—Å—Ç–æ—Ä—ñ—ó –Ω–∞–≤—á–∞–Ω–Ω—è
history = model.fit(X, Y, epochs=10, batch_size=32, verbose=1)

# –ü–µ—Ä–µ–¥–±–∞—á–µ–Ω–Ω—è –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –º–æ–¥–µ–ª—ñ
predictions = model.predict(X)

# –ü–æ–±—É–¥–æ–≤–∞ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(X, Y, label='–î–∞–Ω—ñ')
plt.scatter(X, predictions, color='r', label='–ü—Ä–æ–≥–Ω–æ–∑–∏')
plt.title('–°–ø—ñ–≤—Å—Ç–∞–≤–ª–µ–Ω–Ω—è –î–∞–Ω–∏—Ö —Ç–∞ –ü—Ä–æ–≥–Ω–æ–∑—ñ–≤')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.title('–§—É–Ω–∫—Ü—ñ—è –í—Ç—Ä–∞—Ç')
plt.xlabel('–ï–ø–æ—Ö–∏')
plt.ylabel('–í—Ç—Ä–∞—Ç–∏')

plt.show()
```
